This data set includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family (pp. 500-525). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like 'leaflets three, let it be' for Poisonous Oak and Ivy.

## Carregando e Pré-processando

```python
# Lê os dados com primeira coluna como cabeçalho
dataset = pd.read_csv('data\mushroom.csv',header=0)

# Embaralha os dados
dataset = dataset.sample(frac=1)

# Nomeando colunas
dataset.columns = ['target','cap-shape','cap-surface','cap-color','bruises','odor','gill-attachment','gill-spacing',
             'gill-size','gill-color','stalk-shape','stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring',
             'stalk-color-below-ring','veil-type','veil-color','ring-number','ring-type','spore-print-color','population',
             'habitat']

# Visualizando os dados
dataset.head()
```

**target**: labels para cassificação dos cogumelos entre **e** (edible / comestível) e **p** (poisonous / venenoso)

### Codificando Variáveis Categóricas

Codificando de strings para números (sklearn só aceita números)

```python
for label in dataset.columns:
    dataset[label] = LabelEncoder().fit(dataset[label]).transform(dataset[label])

dataset.head()
```

### Separando X e Y

```python
X = dataset.drop('target',axis=1)
Y = dataset['target'].where(dataset['target']==1,-1)
```

## Treinando e avaliando o modelo

Aqui está sendo treinado apenas um modelo de árvore de classificação do tipo **stump** (1 nível)

```python
tree_model = DecisionTreeClassifier(criterion="entropy",max_depth=1)
predictions = np.mean(cross_validate(tree_model,X,Y,cv=100)['test_score'])

print('Acurácia: ',predictions*100,'%')
```

# Boosting

```python
class Boosting:

    def __init__(self,T):
        self.T = T # número de modelos 
        self.alphas = [] # Guarda os alpha
        self.models = [] # Guarda os modelos 
        self.individual_accuracies = [] # Guarda as acurácias individuais de cada modelo
        self.accuracy = [] # Guarda as acurácias ao longo do treinamento
        self.predictions = None # Guarda as predições 
    
    def fit(self, data):
        # Separa X e Y
        X = data.drop(['target'],axis=1)
        Y = data['target'].where(data['target']==1,-1)

        # Dataframe para cálculo dos erros e pesos 
        Evaluation = pd.DataFrame(Y.copy())
        Evaluation['weights'] = 1/len(data) # Pesos iniciais w = 1/N para todas as observações
        

        # Roda o algoritmo criando cada um dos T modelos (ponderados)
        alphas = [] 
        for t in range(self.T):

            # Define os "Decision Stumps" (max_depth = 1)
            Tree_model = DecisionTreeClassifier(criterion="entropy",max_depth=1) 
            # Treina os stumps utilizando os pesos definidos no passo anterior
            model = Tree_model.fit(X,Y,sample_weight=np.array(Evaluation['weights'])) 
            
            # Guarda o modelo ("weak learner")
            self.models.append(model)
            
            # Calcula as predições do modelo atual
            predictions = model.predict(X)

            # Guarda predições do modelo atual no DF auxiliar
            Evaluation['predictions'] = predictions
            Evaluation['misclassified'] = np.where(Evaluation['predictions'] != Evaluation['target'],1,0)

            # Calcula acurácia e número de observações classificadas erradas
            accuracy = 1 - sum(Evaluation['misclassified'])/len(Evaluation['misclassified'])
            self.individual_accuracies.append(accuracy*100)
            print('Acurácia do modelo {0} individualmente: '.format(t+1),accuracy*100,'%')


            # Calcula o erro (ponderado pelos pesos)
            err = np.sum(Evaluation['weights']*Evaluation['misclassified'])/np.sum(Evaluation['weights'])
 
   
            # Calcula e armazena os multiplicadores alpha ("amount of say")
            alpha = 0.5 * np.log((1-err)/err)
            self.alphas.append(alpha)


            # Atualiza os pesos wi para o treinamento do próximo modelo
            Evaluation['weights'] *= np.exp(alpha*Evaluation['misclassified'])
            
    def predict(self, data):
        # Separa features dos labels
        X_test = data.drop(['target'],axis=1).reindex(range(len(data)))
        Y_test = data['target'].reindex(range(len(data))).where(data['target']==1,-1)
    
        # Faz uma predição com cada modelo 
        predictions = []
        for alpha,model in zip(self.alphas,self.models):
            
            # Pondera a predição utilizando alpha
            prediction = alpha*model.predict(X_test) 
            predictions.append(prediction)
            self.accuracy.append(np.sum(np.sign(np.sum(np.array(predictions),axis=0))==Y_test.values)/len(predictions[0]))

        self.predictions = np.sign(np.sum(np.array(predictions),axis=0))
```

## Treinando modelos com Boosting

```python
number_of_base_learners = 50
model = Boosting(number_of_base_learners)
model.fit(dataset)
```

## Resultados

```python
yhat = model.predict(dataset)
print('Com ',number_of_base_learners,' modelos, a acurácia obtida foi ',model.accuracy[-1]*100,'%')  
```
